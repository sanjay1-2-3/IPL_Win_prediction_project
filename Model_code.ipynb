```python
match = pd.read_csv('matches.csv')
match.head(10)

## Here every row is a match of IPL
```

```python
delivery = pd.read_csv('deliveries.csv')
delivery.head(10)
```

```python
## We don't really need all of this data, we need some selected data to built a predictor
```

```python
total_score_df = delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index() # using reset_index to do the indexing of the data.

# Shows the total score in both the innings of a match

total_score_df
```

```python
total_score_df = total_score_df[total_score_df['inning']==1] ## Taking the total score of 1st innings of all the matches
```

```python
total_score_df
```

```python
# Now we have the 1st inning total of all the matches, so we will merge it with the matches table

match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id') ## not taking the inning number,because we know its 1st inning total

match_df.head()
```

```python
## All the teams of IPL who have played a match during this duration

match['team1'].unique()
```

```python
# Now there are certain teams who were part of IPL, but no longer play so we will remove them 

# Similarly there are teams who have changed name such as
## Delhi Daredevils ------> Delhi Capitals
## Deccan Charger --------> Sunrisers Hyderabad
# So we have to convert the old names to new names in this case.
```

```python
## Created a list of all the active teams in IPL

teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
## Replacing Delhi Daredevils with Delhi Capitals
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

## Replacing Deccan Chargers with Sunreisers Hyderabad
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
## Taking only those teams, which we have in our list of active teams(for both team1 and team2 columns)

match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.head()
```

```python
## Number of rows in match_df has decreased
match_df.shape
```

```python
# Now we don't want to consider those matches where the result was decided by  'Duckworth–Lewis method'
# So, we will take only those rows where dl_applied = 0

match_df=match_df[match_df['dl_applied'] == 0]
match_df.head()
```

```python
match_df.shape

# Number of rows decreased from 841 to 626
```

```python
## Now we will consider only those columns which we need and rest of the columns we can just ignore

match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
match_df.head()
```

```python
## Now we will merge both match and delivery table

delivery_df = match_df.merge(delivery, on='match_id')
delivery_df.head()
```

```python
delivery_df.shape
```

```python
## Now we will filter the data in delivery_df for inning = 2

delivery_df = delivery_df[delivery_df['inning'] == 2]

delivery_df.head()
```

```python
delivery_df.shape
```

```python
## In delivery_df
# (1) total_runs_x ------> Total runs required to win
# (2) total_runs_y ------> Runs scored on every delivery 
```

## Now will find runs left and balls left and wickets left for the team batting second

```python
## Finding current score of team batting second

delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
## Finding runs left 

delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
## Finding balls left


delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])

# Here we are using 126, and not 120 because the over number starts with 1 and not 0.
```

```python
delivery_df.head()
```

```python
# Filling NaN data with 0 in th player_dismissed column

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df
```

```python
# If data != 0 then replace it with 1

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
```

```python
delivery_df
```

```python
# Changing the data type to int64 for player_dismissed column

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
delivery_df
```

```python
# Finding the number of wickets fallen

wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
# Entering the number of wickets left in our delivery_df

delivery_df['wickets'] = 10 - wickets
```

```python
delivery_df
```

```python
delivery_df.tail()
```

## Now we will calculate CRR and RRR for the team batting second

```python
# crr = runs/overs = runs/(number of ball played/ 6) = (runs * 6)/number of balls played
#                                                    = (runs * 6)/(120-number of balls left)

delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df
```

```python
# rrr = runs left/overs left  =   (runs left * 6)/balls left

delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
## Now we have to show result in 0 and 1
# LOGIC:
# if batting and winning team are same then return 1
# if batting and winning team are not same then return 0

def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
delivery_df
```

```python
# Now we have all the data, we just need to takeout the columns that are useful to us, which are:
#     (1) Batting Team
#     (2) Bowling Team
#     (3) City
#     (4) Runs Left
#     (5) Balls Left
#     (6) Wickets Left
#     (7) Total Runs
#     (8) Current Run Rate (CRR)
#     (9) Required Run Rate (RRR)
#     (10) Result


final_df = delivery_df[['batting_team', 'bowling_team','city', 'runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
final_df
```

```python
# Now we will shuffle the data so that there won't be any bias in our model prediction

final_df = final_df.sample(final_df.shape[0])

final_df ##Showing shuffled data
```

```python
## any random row from the shuffled data
final_df.sample()
```

```python
final_df.isnull().sum() # checking if there are null values in our dataframe
```

```python
final_df.dropna(inplace=True) # dropping all the null values in our dataframe by using dropna function with inplace=True(because we want to modify our dataframe)
```

```python
final_df.isnull().sum() ## No null values now
```

```python
# We are not including those rows, where balls left =0, because that row is giving us rrr as +inf or -inf

final_df = final_df[final_df['balls_left'] != 0]
```

## Model preparation

```python
# Spliting the data for training and testing

X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

### Now when we look at the data, then we can see that the data in 'batting_team', 'bowling_team' and 'city' are in strings and are categorical data.

### So we will use OneHotEncoding to replace the data in these 3 columns with numerical data.

```python
## Importing column transformer and onehotencoder to trasform our data in those 3 columns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False, drop='first'),['batting_team', 'bowling_team', 'city'])
]
    ,remainder='passthrough')
```

```python
#Importing Logistic Regression, random forest classifier and pipeline from sklearn

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
## Building our pipeline with two steps
#(1) Transformation
#(2) Logistic Regression Model

pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
# Training our pipeline

pipe.fit(X_train, y_train)
```

```python
## Now we can see that there are no +inf or -inf values for rrr

X_train.describe()
```

```python
## Making the prediction for X_test

y_pred = pipe.predict(X_test)
```

```python
## Importing accuracy score from metrics

from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
# We could have also used score function on our pipeline, they are the same thing.

pipe.score(X_test, y_test)
```

### Here we have the accuracy_score of 80% when we use logistic regression
### When we use random forest classifier for our data then we get the accuracy to be around 99.9%.

## If we had to make the prediction, whether the chasing team will win or not, then we would have used Random forest classifier as it gives results with very high accuracy.

## But we want to show the probability of teams winning at a certain point in match, that is why we will use logistic regression because logistic regression gives us very realistic winning probabilities for both the teams unlike random forest classifier which gives a very extreme probabilites for the teams.

```python
pipe.predict_proba(X_test)[4] # showing us the probability for 5th(i=4) row in X_test i.e KXIP VS CSK 
```
